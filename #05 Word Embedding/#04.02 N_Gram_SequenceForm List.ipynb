```python
text
```

```python
corpus = text.split()

uniq_text = set(corpus)

text_to_int = {}
```

```python
for i, c in enumerate (uniq_text):
    text_to_int.update({c: i})
```

```python
text_to_int
```

```python
data_vec_int = []

for i in range(0,len(corpus)):
    # print(corpus[i])
    data_vec_int.append(text_to_int[corpus[i]])
```

```python
data_vec_int
```

```python
for i in range(0,len(data_vec_int)):
    print(data_vec_int[: i +1])
```
